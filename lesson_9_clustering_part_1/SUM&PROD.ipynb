```python
min_val = 2
max_val = 99

all_pairs = []
for a in range(min_val, max_val + 1):
    for b in range(a, max_val + 1):
        all_pairs.append((a, b))

#чи знає добуток зразу 
def check_dobutok(a, b):
    p = a * b
    count = 0
    for x in range(min_val, max_val + 1):
        for y in range(x, max_val + 1):
            if x * y == p:
                count += 1
    #count > 1 значить не знає
    return count > 1

#Сума
def check_suma_vpevnenist(a, b):
    s = a + b
    for x in range(min_val, s // 2 + 1):
        y = s - x
        if y > max_val: continue
        if check_dobutok(x, y) == False:
            return False
    return True


#
step1_pairs = []
for p in all_pairs:
    if check_suma_vpevnenist(p[0], p[1]):
        step1_pairs.append(p)

# Крок 2: Добуток каже: "Тепер я знаю"
# Значить, серед відфільтрованих пар лише одна має такий добуток
step2_pairs = []
for p in step1_pairs:
    p_val = p[0] * p[1]
    matching_pairs = 0
    for x in step1_pairs:
        if x[0] * x[1] == p_val:
            matching_pairs += 1
    
    if matching_pairs == 1:
        step2_pairs.append(p)

# Крок 3: Сума каже: "Тепер і я знаю"
# Значить, серед тих, що лишилися, лише одна пара має таку суму
final_results = []
for p in step2_pairs:
    s_val = p[0] + p[1]
    matching_pairs = 0
    for x in step2_pairs:
        if x[0] + x[1] == s_val:
            matching_pairs += 1
            
    if matching_pairs == 1:
        final_results.append(p)

# Виводимо результат
print("Результат (числа a та b):", final_results)
```
